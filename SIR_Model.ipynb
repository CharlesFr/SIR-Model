<h1>The SIR Model</h1>

<h3>S = Susceptible</h3>
The group of people who are yet to be infectected

<h3>I = Infected</h3>
The group of people who are current infected

<h3>R = Recovered</h3>
The group of people have recovered and considered immune



```python
%matplotlib inline
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from ipywidgets import interactive
import matplotlib.ticker as mticker

N = 1.0 # number of population
```

```python
I0, R0 = 0.01, 0 # number of people initially infected and recovered
S0 = N - I0 - R0 # people susceptable to infection
beta, gamma = 0.2, 1./10 # Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
```

```python
# differenatial equation
def deriv(y, t, N, beta, gamma): 
    S, I, R = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I
    return dSdt, dIdt, dRdt
```

```python
y0 = S0, I0, R0
maxT = 180
t = np.linspace(0,maxT, maxT) # A grid of time points (in days)

revoveryPeriod, contactPerday = '', ''
```

```python
def UpdatePlot(transRate, daysRecov):
    beta = transRate
    gamma = 1./daysRecov
    ret=odeint(deriv, y0, t, args=(N, beta, gamma))
    S, I, R = ret.T
    R0 = "R0: " + str(round(beta/gamma, 2))
    plt.figure(figsize=(15,5))
    plt.text(174,0.1,R0,bbox=dict(facecolor='w',edgecolor='black',boxstyle='square'))
    plt.plot(t, S, 'b', label='Sceptible')
    plt.plot(t, I, 'r', label='Infeced')
    plt.plot(t, R, 'g', label='Recovered')
    plt.plot(t, I*0.1, 'k', label='Needing Hospitalisation')
    plt.plot([0, maxT], [0.00225754, 0.00225754], '--', lw=1, color="0.5", label='Total Hospital Beds')
    plt.legend(loc='best')
    plt.xlabel('Days')
    plt.xlabel('Fraction of Polulation')
    plt.grid()
    
    
```

```python
interactive_plot = interactive(UpdatePlot, transRate=(0.01, 0.5, 0.001), daysRecov=(4, 30, 1))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```
